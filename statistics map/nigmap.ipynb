```python
import json 
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default ="browser"
```

```python

with open(r"C:\Users\tomis\Desktop\NBS_states.geojson") as f:
        geo_data = json.load(f)
```

```python
state_id_map={}
for feature in geo_data['features']:
    feature['id']=feature['properties']['OBJECTID']
    state_id_map[feature['properties']['OBJECTID']]= feature['id']
```

```python

pop_df=pd.read_csv('states_nigeria.csv', sep= ',' )
geo_data['features'][1]['properties']  
```

```python
fig = px.line(pop_df, x = 'State', y ='Underemployment Population', title='Underemployment Population against State (2020)')
fig.show()
```

```python
fig = px.line(pop_df, x = 'State', y ='Rate', title='Underemployment Rate against State (2020)')
fig.show()
```

```python
nig_coords= {"lat": 9.0820, "lon": 8.6753}
```

```python
 fig= px.choropleth_mapbox(
     pop_df,
     locations= "SNo",
     geojson=geo_data,
     color="Underemployment Population",
     hover_name="State",
     center= nig_coords,
     zoom = 5,
     opacity = 0.75,
     color_continuous_scale= "mint",
     mapbox_style="carto-positron",
     labels = {"Underemployment Population", "State"}
 )
fig.update_layout(margin ={"t":0,"b":0,"l":0,"r":0})

fig.show()
```
